```python
def minmax(lis):
        
    Min = min(lis)
    Max = max(lis)
    
    l =np.array([round((x-Min)/(Max-Min)*(1-0)+0,3) for x in lis])
        
    return l
```

```python
import numpy as np

X = np.array([1100,1400,1425,1550,1600,1700,1700,1875,2350,2450])
Y = np.array([199,245,319,240,312,279,310,308,405,324])

Xminmax = minmax(X)
Yminmax = minmax(Y)
```

```python
for i in range(len(X)):
    print(Xminmax[i],"\t",Yminmax[i])
    

    np.set_printoptions(precision=5)
    
```

```python
def gradientdescent(X,Y,a=0.45,b=0.75,r=0.01):
    
    x=0
    
    flag = 1
    
    while flag:
        Ypred = a + (b*Y)
        
        SSE = ((Ypred-Y) ** 2 )/ 2 
        
        SSE_total = np.sum(SSE)
        
        dowA = -(Y-Ypred)
        dowB = -(Y-Ypred)*X
        
        Aerr_tot = np.sum(dowA)
        Berr_tot = np.sum(dowB)
        
        newA = abs(a - r*Aerr_tot)
        newB = abs(b - r*Berr_tot)
    
        x+=1
    
        if float(SSE_total) < 0.05 or x > 100000 or float(newA) < r/100000000 or float(newB) < r/100000000: flag=0
        else :
            a = (newA);
            b = (newB);
    
    print("SSE_total","a","b","x")
    print(round(SSE_total,10),round(a,10),round(b,10),x,"\n")
        
```

```python
x = 0

from random import randint

while x<100 :
    a = randint(1,101)/100
    b = randint(1,101)/100

    print(a,b)
    
    gradientdescent(Xminmax,Yminmax,a,b)
    x += 1
```
